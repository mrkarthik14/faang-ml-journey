```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def vector_addition(v,w):
        """ 
        Adds two different vectors element vise
        v is one vector and it represnted using np.array
        w is another vector and it is also represented using np.array
        return the sum of two vectors
        """
        return v + w
        
```

```python
v = np.array([2,3])
w = np.array([4,-1])
result = vector_addition(v,w)
print("Result of vector addition:", result)
```

```python
def vector_addition(v, w):
    if len(v) != len(w):
        raise ValueError("Vectors must be same length")

    v1 = []
    for i in range(len(v)):
        add = v[i] + w[i]
        v1.append(add)

    return v1
print(vector_addition(v,w))
```

```python
origin = np.array([0,0])
plt.quiver(*origin, v[0],v[1],color ='red', angles ='xy', scale_units='xy', scale=1,label = 'v')
plt.quiver(*origin, w[0], w[1], color = 'green', angles='xy', scale_units ='xy', scale=1, label = 'w')
plt.quiver(*origin, result[0], result[1], color = 'blue', angles='xy', scale_units = 'xy', scale=1, label = 'v+w')
plt.xlim(-1,7)
plt.ylim(-2,5)
plt.text(v[0], v[1], 'v', color='red')
plt.text(w[0], w[1], 'w', color='green')
plt.text(result[0], result[1], 'v + w', color='blue')
plt.legend()
plt.grid()
plt.show()
```

# Vector Multiplication

### 1. Scalar Multiplication

```python
# multiplying vector by scalar
scalar = 3.5

vector = np.array([2,3])
scaled_vector = scalar * vector
print("Scaled vector:", scaled_vector)
```

```python
plt.quiver(*origin, vector[0], vector[1],
        color='green',
        angles='xy', scale_units='xy',
        scale=1, label='Original Vector',
        zorder = 3)

plt.quiver(*origin, scaled_vector[0], scaled_vector[1],
        color='red',
        angles='xy', scale_units='xy',
        scale=1, label='Scaled Vector',
        zorder = 2)

plt.text(vector[0], vector[1]/1.15, 
        'Original Vector', color='green',
        zorder = 4)

plt.text(scaled_vector[0], scaled_vector[1]/1.065, 
        'Scaled Vector', color='red',
        zorder = 1)

plt.xlim(0,15)
plt.ylim(0,15)
plt.grid()
plt.legend()
plt.show()
```

### 2. Vector Magnitude

```python
# magnitude of vector
vec = np.array([3,4])

def comput_mag(vec):
        s =0
        for i in vec:
                s += i**2
        magnitude = np.sqrt(s)
        return magnitude

# magnitude = np.sqrt(vec[0]**2 + vec[1]**2)
# magnitude = np.linalg.norm(vec)
magnitude = comput_mag(vec)
print("Magnitude of the vector:", magnitude)
```
