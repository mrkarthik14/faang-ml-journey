### Class Variables
Shared among all instances (objects) of a class <br>
Defined outside the constructor <br>
Allow you to share data among all objects created from that class

```python
class Jackie_chan_tv_show:
        # CLASS VARIABLES
        release = 2000
        actors =0
        def __init__(self,name, age, gender, location):

                # INSTANCE VARIABLES
                self.name = name
                self.age =age
                self.gender = gender
                self.location = location

                # we can use hero,main_lead instead of self
                # when we want specifically
                # we can use class variables
                Jackie_chan_tv_show.actors +=1

# objects = class name (class parameters/arguments
hero = Jackie_chan_tv_show('Jackie Chan', 34, 'Male','Japan')
main_lead = Jackie_chan_tv_show('Julie', 12, 'Female','Tokyo')

# acces the class variables using class name 
print(f' Jackie_chan_Tv_Show_Hero : {hero.name}\n release year : {Jackie_chan_tv_show.release}')

# we print the class varibles
print(f" actors : {Jackie_chan_tv_show.actors}")
```

##### Inheritance
Allows a class to inherit attributes and methods from another class <br>
Helps with code reusability and extensibility<br>
class Child(Parent)

```python
# employee is a parent class 
#
class Employee:
        def __init__(self,name, salary):
                self.name = name
                self.salary = salary

        def get_details(self):
                return f'{self.name} earns {self.salary}'
        
class DataAnalyst(Employee):
        def get_role(self):
                return 'Data Analyst'
        
da = DataAnalyst('Karthik',600000)
print(da.get_details())
print(da.get_role())

```

```python
class DataScientist(Employee):
        def get_details(self):
                return f'{self.name} earns {self.salary} and builds ML models'
        
ds = DataScientist('Charan', 9000000)
print(ds.get_details())
```
