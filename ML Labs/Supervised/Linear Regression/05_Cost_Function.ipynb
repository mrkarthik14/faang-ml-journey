## Cost Function (Mean Squared Error - MSE)


#### Definition
The Cost Function (or Loss Function) measures how "wrong" the model is on the training data. For linear regression, the standard cost function is Mean Squared Error (MSE).

#### Mathematical Formula

 $$J(W, B) = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}_i - y_i)^2$$

##### Where:

m: Number of training samples       

Ŷᵢ: Model prediction for sample i

yᵢ: True label for sample i

J(W, B): Total cost (depends on W and B)

#### Breaking It Down

$$(\hat{y}_i - y_i)^2$$

Difference between prediction and truth

Squared: penalizes large errors more, ensures positive value
 
 $$\frac{1}{m} \sum_{i=1}^{m}$$

Average over all samples
Normalizes to per-sample basis
Makes cost independent of dataset size

#### Intuition

Lower J = Better model (predictions closer to truth)

Higher J = Worse model (predictions far from truth)

J = 0 means perfect predictions

Goal of training: Minimize J by finding best W and B

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Training Data
x = np.array([600, 800, 1000, 1200, 1400, 1600, 1800]).reshape(-1,1)
y = np.array([180000, 200000, 230000, 255000, 280000, 305000, 330000])
```

```python
# Train Model
model = LinearRegression()
model.fit(x,y)
y_pred = model.predict(x)
```

```python
# method1: using sklearn
mse_sklearn = mean_squared_error(y, y_pred)
print(f"MSE (sklearn): {mse_sklearn:.2f}")
```

```python
# method 2 : manual calculation
residuals = y - y_pred.flatten()
mse_manual  = np.mean(residuals ** 2)
print(f"MSE Manual: {mse_manual}")
```

```python
# method 3 : step by step breakdown

for i in range(len(x)):
    pred = y_pred[i]
    actual = y[i]
    residual = actual - pred
    squared_error = residual ** 2

    print(f"sample : {i+1}")
    print(f"    Actual : ${actual:,.0f}")
    print(f"    Predicted : ${pred:,.0f}")
    print(f"    Residual : ${residual:,.0f}")
    print(f"    Squared error : ${squared_error:,.0f}")

sum_squared_errors =  np.sum(residuals ** 2)
mse = sum_squared_errors / len(x)

print(f"\n sum of squared errors: ${sum_squared_errors:,.0f}")
print(f"number of sample : {len(x)}")
print(f"MSE = {sum_squared_errors:,.0f} / {len(x)} = ${mse:,.0f}")
```

## Visualizing the Cost Function

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# ------------------------------------------------------------
# 1. Small dataset to illustrate how cost varies with W and B
# ------------------------------------------------------------
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
```

```python
# ------------------------------------------------------------
# 2. Create a grid of possible W and B values (parameter space)
#    This simulates scanning many models to see their cost.
# ------------------------------------------------------------
w_range = np.linspace(-1,3,50)
b_range = np.linspace(-1,3, 50)
w_grid, b_grid  = np.meshgrid(w_range, b_range)
```

```python
# ------------------------------------------------------------
# 3. Compute the cost J(W, B) for each point in the grid
#    J = Mean Squared Error for predictions on the dataset
# ------------------------------------------------------------
J = np.zeros_like(w_grid)

for i in range(len(w_range)):
    for j in range(len(b_range)):
        
        # pick the W and B at this point on the grid
        w_test = w_grid[j,i]
        b_test = b_grid[j,i]
        
        # compute predictions for all X values
        y_pred = w_test * x + b_test
        
        # compute cost (MSE) for this model
        J[j, i] = np.mean((y_pred - y) ** 2)
```
