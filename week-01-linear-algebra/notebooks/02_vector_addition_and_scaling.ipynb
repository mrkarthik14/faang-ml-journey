### 1. Scalar Multiplication


- Scales vector magnitude
- Direction unchanged (if scalar > 0)
- In ML: learning rate × gradient


```python
# multiplying vector by scalar
def scalar_multiply(c,v):
        return [c* x for x in v]

scalar = 4
scaled_v = scalar_multiply(scalar,v)
vector = np.array([2,3])
scaled_vector = scalar * vector
print("Scaled vector:", scaled_vector)
```

### 2. Vector Magnitude


Magnitude represents vector length.

Formula:
||v|| = sqrt(x² + y²)


```python
# magnitude of vector
vec = np.array([3,4])

def comput_mag(vec):
        s =0
        for i in vec:
                s += i**2
        magnitude = np.sqrt(s)
        return magnitude
        # return math.sqrt(sum(x**2 for x in v))

# magnitude = np.sqrt(vec[0]**2 + vec[1]**2)
# magnitude = np.linalg.norm(vec)
magnitude = comput_mag(vec)
print("Magnitude of the vector:", magnitude)
```
